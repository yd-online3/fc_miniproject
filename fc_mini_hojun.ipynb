### format 열을 기준으로 PAYMENT, REFUND 로 나누어 환불율을 계산 

```python
# 환불이 승인된 데이터프레임
cond_refund = (df['type']== 'REFUND') & (df['state'] == 'COMPLETED')

# 결제가 승인된 데이터 프레임
cond_payment = (df['type']== 'PAYMENT') & (df['state'] == 'COMPLETED')

# format array화
format_array = df['format'].unique()
format_refund_rate = []

for i in format_array:
    cond = (df['format'] == i)
    # 결제승인
    df[cond][cond_payment]
    # 환불승인
    df[cond][cond_refund]
    # 결제승인 갯수 = 전체 갯수
    payment = len(df[cond][cond_payment].index)
    # 환불승인 갯수
    refund = len(df[cond][cond_refund].index)
    if refund != 0 and payment != 0:
        refund_ratio = refund / payment * 100
    elif payment == 0:
        refund_ratio = 0
        print(f'format : {i}의 payment가 0 입니다.')
    refund_ratio = round(refund_ratio, 3)
    print(f'format : "{i}"의 환불률은 {refund_ratio}% 입니다.')
    format_refund_rate.append(refund_ratio)
format_refund_rate = np.array(format_refund_rate)
```

```python
# PAYMENT 금액 가져오기
df3 = df

cond = df3[df3['state']=='REQUESTED'].index
df3 = df3.drop(cond)

cond = df3[df3['state']=='프로모션'].index
df3 = df3.drop(cond)

cond = df3[df3['type']=='REFUND'].index
df3 = df3.drop(cond)
payment_list = []
how5 = {
    'transaction_amount' : np.sum,
}
df3.groupby(['format','type','state']).agg(how)
for i in range(14):
    payment_list.append(df3.groupby(['format','type','state']).agg(how).iloc[i,0])
payment_list.insert(9,0)

# REFUND 금액 가져오기
df4 = df

cond = df4[df4['state']=='REQUESTED'].index
df4 = df4.drop(cond)

cond = df4[df4['state']=='프로모션'].index
df4 = df4.drop(cond)

cond = df4[df4['type']=='PAYMENT'].index
df4 = df4.drop(cond)
df4.groupby(['format','type','state']).agg(how)
refund_list = []
how5 = {
    'transaction_amount' : np.sum,
}
df4.groupby(['format','type','state']).agg(how)
for i in range(13):
    refund_list.append(df4.groupby(['format','type','state']).agg(how).iloc[i,0])
refund_list.insert(3,0)
refund_list.insert(5,0)

# 그래프용 데이터 df5
df5 = pd.DataFrame(
    data = {'PAYMENT' : payment_list, 
            'REFUND' : refund_list},
    index = np.sort(df['format'].unique(), axis=0),
    columns=['PAYMENT','REFUND']
)
df5['transaction_amount_refund_rate'] = round((abs(df5['REFUND']) / (df5['PAYMENT'] ) *100), 2)
df5['format_refund_rate'] = np.round(format_refund_rate, 2)

df5
```

## 각 포맷마다 

```python
# df5 필요한 항목들을 -> array로 변환
format_array = np.array(df5.index)
format_refund_rate = np.array(df5['format_refund_rate'])
transaction_amount_refund_rate = np.array(df5['transaction_amount_refund_rate'])

fig = plt.figure(figsize=(14, 8))

ax = fig.add_subplot(111)
lns1 = ax.plot(format_array, format_refund_rate, color = 'red',  label='Format Refund Rate')
ax.set_ylabel('Format Refund Rate(%)')
plt.xticks(rotation=-30)

# ax2는 y2에 대한 그래프, twinx로 x축을 공유
ax2 = ax.twinx()
lns2 = ax2.plot(format_array, transaction_amount_refund_rate, color = 'blue', label='Transaction Amount Refund Rate')
ax2.set_ylabel('Transaction Amount Refund Rate(%)')
ax2.grid()

# 범뢰 같이 표현
lns = lns1+lns2
labs = [l.get_label() for l in lns]
ax.legend(lns, labs, loc=0)
plt.show()
```
